```python
import numpy as np
import tensorflow as tf
```

```python
# model 생성
model=tf.keras.Sequential([
  tf.keras.layers.Dense(2, activation='tanh'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# weight update 시 sgd 알고리즘 사용, learning rate=0.15 설정
sgd = tf.keras.optimizers.SGD(learning_rate=0.2)
model.compile(optimizer=sgd, loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
X=np.array([[0,0], [0,1], [1,0], [1,1]])
T=np.array([0, 1, 1, 0])
```

```python
# model 학습
model.fit(X, T, epochs=20000, verbose=0)
```

```python
# 만들어진 model의 architecture
model.summary()
```

```python
# predict 결과 출력
print(model.predict(X))
```
